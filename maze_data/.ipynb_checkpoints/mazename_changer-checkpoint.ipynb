```python
import glob 
import os
from natsort import natsorted
paths = glob.glob('./*.txt')
cnt = 49
for path in natsorted(paths):
    cnt += 1
    if cnt == 100:
        cnt = 50
    print(cnt,path,path[:path.index('_')]+'_'+str(cnt)+'.txt')
    os.rename(path, path[:path.index('_')]+'_'+str(cnt)+'.txt')
```
